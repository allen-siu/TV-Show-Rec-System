```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
anime_csv_path = './anime.csv'
ratings_csv_path = './rating.csv'
```

```python
anime_data = pd.read_csv(anime_csv_path)
anime_data['genre'].fillna('', inplace=True)
anime_data.head()

# Use this line to check for Nan genre value
# anime_data.iloc[2844]
```

```python
anime_data['genre_array'] = anime_data['genre'].str.split(', ')
anime_data.head()
```

```python
for index, anime in anime_data.iterrows():   
    for genre in anime['genre_array']:
        if f'is_{genre}' not in anime_data.columns.tolist():
        
            anime_data[f'{genre}'] = anime_data['genre_array'].apply(
                lambda genre_array: 1 if genre in genre_array else 0
            )

anime_data.drop(columns=['genre_array'], inplace=True)
anime_data.head()
```
